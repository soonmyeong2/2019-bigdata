```python
import pandas as pd

df = pd.read_csv("ACCOMMODATION.csv")
resultDf = pd.DataFrame
```

```python
for i in range(len(df)):
    rowDf = pd.DataFrame({"Address":[df.loc[i,"도로명주소"]]})
    resultDf = resultDf.append(rowDf, ignore_index = True)
    
```

```python
print(resultDf)

resultDf.to_csv("ACCOMMODATION.csv", index = False)
```
