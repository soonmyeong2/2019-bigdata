```python
import pandas as pd

yearMonth = 201701

dict = {}

for year in range(3):
    for month in range(12):
        if yearMonth == 201904:
            break
        fileName = "TEMP_FOREIGNER_DONG_{yearmonth}.csv".format(yearmonth = yearMonth)    
        df = pd.read_csv(fileName)
        df = df.reset_index()
        
        for index in range(len(df)):
            dong = df.iloc[index, 2]
            total = df.iloc[index, 3]
            chinese = df.iloc[index, 4]
            notChinese = df.iloc[index, 5]
            if dong in dict:
                dict[dong] = (str(int(dict[dong][0]) + int(total)), str(int(dict[dong][1]) + int(chinese)), str(int(dict[dong][2]) + int(notChinese)))
            else:
                
                dict[dong] = (total, chinese, notChinese)
        
        
        
        yearMonth = yearMonth + 1
    yearMonth = yearMonth + 100 -12

    
```

```python
resultDf = pd.DataFrame()
for key in dict:
    rowDf = pd.DataFrame({"Dong": [key], "TotalPopulation":[dict[key][0]], "Chinese":[dict[key][1]], "NotChinese":[dict[key][2]] })
    resultDf = resultDf.append(rowDf, ignore_index = True)
    
resultDf.to_csv("TEMP_FOREIGNER_DONG.csv")
```
