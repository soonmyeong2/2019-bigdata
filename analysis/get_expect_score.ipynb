```python
from pyspark.sql import SQLContext
from pyspark.sql.types import DoubleType
sqlContext = SQLContext(sc)
```

```python
df = sqlContext.read.csv('data.csv')
df = df.withColumn('_c0', df['_c0'].cast(DoubleType()))
df = df.withColumn('_c1', df['_c1'].cast(DoubleType()))
df = df.withColumn('_c2', df['_c2'].cast(DoubleType()))
df = df.withColumn('_c3', df['_c3'].cast(DoubleType()))
df.show()
```

```python
df = df.withColumn('e', df['_c0'] * 0.07814779 + df['_c1'] * 0.06709774 + df['_c2'] * 0.868377 -8.1491664e-05)
df.show()
```

```python
df_rec = df.withColumn('recommend', df['e'] - df['_c3'])
df_rec.show()
```

```python
maxValue = df_rec.agg({'recommend': 'max'}).collect()[0][0]
maxValue
```

```python
df_norm = df_rec.withColumn('norm', df_rec['recommend'] / maxValue)
df_norm.show()
```

```python
df_norm.repartition(1).write.csv('hdfs://master:9000/user/hadoop/recommend_score')
```
