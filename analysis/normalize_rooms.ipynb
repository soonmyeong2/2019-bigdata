```python
from pyspark.sql import SQLContext
sqlContext = SQLContext(sc)
```

```python
from pyspark.sql.types import IntegerType
```

```python
df = sqlContext.read.csv('hdfs://master:9000/user/hadoop/rooms/part-00000')
df.show()
```

```python
df = df.withColumn('_c1', df['_c1'].cast(IntegerType()))
df.show()
```

```python
maxValue = df.agg({'_c1': 'max'}).collect()[0][0]
maxValue
```

```python
df_norm = df.withColumn('norm', df['_c1'] / maxValue)
df_norm.show()
```

```python
df_result = df_norm[['_c0', 'norm']]
df_result.show()
```

```python
df_result.repartition(1).write.csv('hdfs://master:9000/user/hadoop/rooms_norm')
```
