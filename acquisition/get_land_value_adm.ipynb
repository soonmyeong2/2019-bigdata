```python
from pyspark.sql import SQLContext
sqlContext = SQLContext(sc)
```

```python
df = sqlContext.read.csv('hdfs://master:9000/user/hadoop/merge_LandValue_')
```

```python
df.show()
```

```python
from pyspark.sql.types import IntegerType
from pyspark.sql.functions import substring

df = df.withColumn('jibun', df['_c0']) \
        .withColumn('dong', substring(df['_c0'], 1, 10).cast(IntegerType())) \
        .withColumn('mount', substring(df['_c0'], 11, 1).cast(IntegerType()) - 1) \
        .withColumn('main', substring(df['_c0'], 12, 4).cast(IntegerType())) \
        .withColumn('sub', substring(df['_c0'], 16, 4).cast(IntegerType()))
df.show()
```

```python
df = df.withColumn('value', df['_c1'])[['jibun', 'dong', 'mount', 'main', 'sub', 'value']]
df.show()
```

```python
df_jibun = sqlContext.read.option("delimiter", "|").csv('hdfs://master:9000/user/hadoop/jibun.txt', encoding="euc-kr")
```

```python
df_jibun.show()
```

```python
df_reduce = df_jibun[['_c0', '_c2', '_c7', '_c8', '_c9']]
df_reduce.show()
```

```python
df_joined = df.join(df_reduce, (df['dong'] == df_reduce['_c2']) & (df['mount'] == df_reduce['_c7']) & (df['main'] == df_reduce['_c8']) & (df['sub'] == df_reduce['_c9']))
df_joined.show()
```

```python
df_coded = df_joined.withColumn('code', df_joined['_c0'])[['jibun', 'dong', 'code', 'value']]
df_coded = df_coded.filter(df_coded['code'].isNotNull())
df_coded.show()
```

```python
df_add = sqlContext.read.option("delimiter", "|").csv('hdfs://master:9000/user/hadoop/additional.txt', encoding="euc-kr")
df_add.show()
```

```python
df_mapped = df_coded.join(df_add, df_coded['code'] == df_add['_c0'])
df_mapped = df_mapped.withColumn('adm_code', df_mapped['_c1'])[['jibun', 'value', 'adm_code']]
df_mapped = df_mapped.filter(df_mapped['adm_code'].isNotNull())
df_mapped.show()
```

```python
df_mapped.repartition(1).write.csv("hdfs://master:9000/user/hadoop/land_value_adm")
```
