```python
from pyspark.sql import SQLContext
sqlContext = SQLContext(sc)
```

```python
df = sqlContext.read.csv('hdfs://master:9000/user/hadoop/data/rooms.csv', encoding='utf-8', header=True)
df.show()
```

```python
df = df.na.fill({'한실수': 0, '양실수': 0})
df.show()
```

```python
df_non_null = df[df['도로명주소'].isNotNull()]
df_non_null.show()
```

```python
from pyspark.sql.types import StringType
from pyspark.sql.functions import udf
import requests
import json

def getCode(addr):
    res = requests.get('https://dapi.kakao.com/v2/local/search/address.json', \
                       params={'query': ' '.join(addr.split(' ')[:-1])}, \
                       headers={'Authorization': 'KakaoAK ddaaef4912b0e162f6c1e2edf20b8ddf'})
    info = json.loads(res.text)
    if len(info['documents']) > 0:
        return info['documents'][0]['address']['h_code']
    else:
        return None

get_udf = udf(getCode, StringType())
df_coded = df_non_null.withColumn('adm_code', get_udf(df_non_null['도로명주소']))
df_coded.show()
```

```python
df_coded_null = df_coded[df_coded['adm_code'].isNull()]
df_coded_null[['도로명주소']].show(20, False)
```

```python
def getCode2(addr):
    print(addr)
    slices = addr.split(' ')
    if len(slices) < 4:
        return None
    res = requests.get('https://dapi.kakao.com/v2/local/search/address.json', \
                       params={'query': ' '.join(slices[:4])}, \
                       headers={'Authorization': 'KakaoAK ddaaef4912b0e162f6c1e2edf20b8ddf'})
    info = json.loads(res.text)
    if len(info['documents']) > 0:
        return info['documents'][0]['address']['h_code']
    else:
        return None

get_udf2 = udf(getCode2, StringType())
df_coded2 = df_coded_null.withColumn('adm_code', get_udf2(df_coded_null['도로명주소']))
df_coded2 = df_coded2[df_coded2['adm_code'].isNotNull()]
df_coded2.show()
```

```python
df_union = df_coded[df_coded['adm_code'].isNotNull()].union(df_coded2)
df_union.show()
```

```python
from pyspark.sql.types import IntegerType
```

```python
df_calc = df_union.withColumn('rooms', (df_union['한실수'] + df_union['양실수']).cast(IntegerType()))
df_calc.show()
```

```python
df_result = df_calc.groupBy('adm_code').sum('rooms')
df_result.show()
```

```python
df_result.repartition(1).write.csv('hdfs://master:9000/user/hadoop/rooms')
```
